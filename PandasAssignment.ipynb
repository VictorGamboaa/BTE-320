````python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams.update({'font.size': 20, 'figure.figsize': (10, 8)})
````

````python
import pandas as pd
import matplotlib.pyplot as plt



df = pd.read_csv('travel-times.csv')

df['Date'] = pd.to_datetime(df['Date'])

#Shape of Data Set
print(f'Shape of the dataset: {df.shape}')

#First 15 rows of dataset
print("First 15 rows of the dataset:")
print(df.head(15))

#Information with missing values
missing_values_info = df.isnull().sum()
print("Features with missing values:")
print(missing_values_info[missing_values_info > 0])

#Dropping duplicates
df.drop_duplicates(keep='last', inplace=True)

#Calculating total number of missing values
total_missing_values = df.isnull().sum().sum()
print("Total number of missing values:", total_missing_values)

#Creating 2 copies,dropping rows and columns with missing values
copy1 = df.copy()
copy2 = df.copy()

copy1.dropna(axis=0, inplace=True)
copy2.dropna(axis=1, inplace=True)

#Getting summary statistics
print("Summary statistics:")
print(df.describe())

#Showing rows 11-14
print("Rows 11 to 14:")
print(df.iloc[11:15])

#Subsetwith trips from November to January
subset = df[(df['Date'] == '2011-11-23') | (df['Date'] == '2012-01-06')]

#Scatter plot and Box Plots
plt.scatter(df['Distance'], df['TotalTime'])
plt.xlabel('Distance')
plt.ylabel('TotalTime')
plt.title('Distance and TotalTime')
plt.rcParams.update({'font.size': 20, 'figure.figsize': (10, 8)})
plt.show()


plt.boxplot(df['AvgSpeed'])
plt.xlabel('AvgSpeed')
plt.title('Average Speed')
plt.rcParams.update({'font.size': 20, 'figure.figsize': (10,8)})
plt.show()




plt.boxplot(df['AvgMovingSpeed'])
plt.xlabel('AvgMovingSpeed')
plt.title('Average Moving Speed')
plt.rcParams.update({'font.size': 20, 'figure.figsize': (10, 8)})
plt.show()


````
